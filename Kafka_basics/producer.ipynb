```python
from kafka import KafkaProducer
import json
import random
```

```python
producer1 = KafkaProducer(
    bootstrap_servers= "kafka:9092",
    value_serializer= lambda x: json.dumps(x).encode("utf-8")
)
```

```python
producer2 = KafkaProducer(
    bootstrap_servers= "kafka:9092",
    value_serializer= lambda x: json.dumps(x).encode("utf-8")
)
```

```python
for i in range(10):
    producer1.send(
        "demo-topic",
        {
            "id": i,
            "name": f"Student-{i}"
        }
    )

producer1.flush()
print("Done")
```

```python
for i in range(101, 111):
    producer2.send(
        "demo-topic",
        {
            "id": i,
            "order": random.choice([1,2,3,4,5])
        }
    )
producer2.flush()
print("Done")
```
